```python
import os
import PIL
import tkinter as tk
from tkinter import ttk
from pycaw.magic import MagicApp
from pycaw.constants import AudioSessionState
from pycaw.pycaw import AudioUtilities, ISimpleAudioVolume
```

```python
def handle_all(*args):
    print("callback")
    print(args)

def toggle_mozilla():
    global mozilla_checked
    mozilla_checked = not mozilla_checked
    canvas.itemconfigure(mozilla, fill="green" if mozilla_checked else "gray")

def toggle_edge():
    global edge_checked
    edge_checked = not edge_checked
    canvas.itemconfigure(edge, fill="green" if edge_checked else "gray")

def toggle_whatsapp():
    global whatsapp_checked
    whatsapp_checked = not whatsapp_checked
    canvas.itemconfigure(whatsapp, fill="green" if whatsapp_checked else "gray")

def toggle_chrome():
    global chrome_checked
    chrome_checked = not chrome_checked
    canvas.itemconfigure(chrome, fill="green" if chrome_checked else "gray")

def toggle_vlc():
    global vlc_checked
    vlc_checked = not vlc_checked
    canvas.itemconfigure(vlc, fill="green" if vlc_checked else "gray")

def toggle_discord():
    global discord_checked
    discord_checked = not discord_checked
    canvas.itemconfigure(discord, fill="green" if discord_checked else "gray")

def toggle_windowsvideo():
    global windowsvideo_checked
    windowsvideo_checked = not windowsvideo_checked
    canvas.itemconfigure(windowsvideo, fill="green" if windowsvideo_checked else "gray")

def toggle_netflix():
    global netflix_checked
    netflix_checked = not netflix_checked
    canvas.itemconfigure(netflix, fill="green" if netflix_checked else "gray")

def service():
    # Find the target application by its name
    target_app_names = []
    spotify_app_name = "Spotify.exe"

    global started
    started = not started

    global mozilla_checked
    global edge_checked
    global whatsapp_checked
    global chrome_checked
    global vlc_checked
    global discord_checked
    global windowsvideo_checked
    global netflix_checked

    textbox_values = textbox.get()
    other_apps = textbox_values.split(",")

    if mozilla_checked:
        target_app_names.append("firefox.exe")
    if edge_checked:
        target_app_names.append("msedge.exe")
    if whatsapp_checked:
        target_app_names.append("WhatsApp.exe")
    if chrome_checked:
        target_app_names.append("chrome.exe")
    if vlc_checked:
        target_app_names.append("vlc.exe")
    if discord_checked:
        target_app_names.append("Discord.exe")
    if windowsvideo_checked:
        target_app_names.append("Video.UI.exe")
    if netflix_checked:
        target_app_names.append("Netflix.exe")
    if other_apps:
        target_app_names.extend(other_apps)
    
    spotify = MagicApp(
        {spotify_app_name},
        volume_callback=handle_all,
        mute_callback=handle_all,
        state_callback=handle_all,
        session_callback=handle_all,
    )

    while started:
        sessions = AudioUtilities.GetAllSessions()
        any_target_app_is_playing = False

        for session in sessions:
            if session.Process and session.Process.name() in target_app_names:
                volume = session._ctl.QueryInterface(ISimpleAudioVolume)
                is_sound = session.State == AudioSessionState.Active
                if is_sound:
                    any_target_app_is_playing = True
                    break
                else:
                    any_target_app_is_playing = False

        if not any_target_app_is_playing:
            if spotify.state:
                spotify.volume = 0.99
        else:
            if spotify.state:
                spotify.volume = 0.3
    

width = 800
height = 500

root = tk.Tk()
root.minsize(width=width, height=height) 
root.maxsize(width=width, height=height) 
root.title("Full Focus But Don't Froget Music")

background_image = tk.PhotoImage(file="src/background.png")  # Replace with your background image path
spotify_image = tk.PhotoImage(file="src/spotify.png")  # Replace with your background image path

# Create a Canvas with a background image
canvas = tk.Canvas(root, width=width, height=height, borderwidth=0, highlightthickness=0)
canvas.pack()

back = canvas.create_image(0, 0, image=background_image, anchor="nw")
spoti = canvas.create_image(450, 425, image=spotify_image, anchor="w")
text = canvas.create_text(20, 150, text="This sweet app turns down the\nvolume of Spotify when there is\nany sound from the selected apps.\nEven when your computer is on fire.\nYou stay focused.\nWe set volume for you.", fill="white", font=("Arial", 12), anchor="nw")

# Create the checkbox item on the Canvas
mozilla = canvas.create_text(317, 142, text="◼", font=("Arial", 20), fill="gray", anchor="w")
edge = canvas.create_text(420, 142, text="◼", font=("Arial", 20), fill="gray", anchor="w")
whatsapp = canvas.create_text(523, 142, text="◼", font=("Arial", 20), fill="gray", anchor="w")
chrome = canvas.create_text(626, 142, text="◼", font=("Arial", 20), fill="gray", anchor="w")

vlc = canvas.create_text(317, 225, text="◼", font=("Arial", 20), fill="gray", anchor="w")
discord = canvas.create_text(420, 225, text="◼", font=("Arial", 20), fill="gray", anchor="w")
windowsvideo = canvas.create_text(523, 225, text="◼", font=("Arial", 20), fill="gray", anchor="w")
netflix = canvas.create_text(626, 225, text="◼", font=("Arial", 20), fill="gray", anchor="w")

# Bind a click event to the checkbox item
canvas.tag_bind(mozilla, "<Button-1>", lambda event: toggle_mozilla())
canvas.tag_bind(edge, "<Button-1>", lambda event: toggle_edge())
canvas.tag_bind(whatsapp, "<Button-1>", lambda event: toggle_whatsapp())
canvas.tag_bind(chrome, "<Button-1>", lambda event: toggle_chrome())
canvas.tag_bind(vlc, "<Button-1>", lambda event: toggle_vlc())
canvas.tag_bind(discord, "<Button-1>", lambda event: toggle_discord())
canvas.tag_bind(windowsvideo, "<Button-1>", lambda event: toggle_windowsvideo())
canvas.tag_bind(netflix, "<Button-1>", lambda event: toggle_netflix())
canvas.tag_bind(spoti, "<Button-1>", lambda event: service())

textbox = tk.Entry(canvas)
textbox_window = canvas.create_window(490, 285, anchor='nw', window=textbox)
label = canvas.create_text(317, 285, text="Another Application: ", font=("Arial", 14), fill="white", anchor="nw")

# Initialize the checkbox state
mozilla_checked = False
edge_checked = False
whatsapp_checked = False
chrome_checked = False
vlc_checked = False
discord_checked = False
windowsvideo_checked = False
netflix_checked = False
started = False


canvas.itemconfigure(text, stipple="gray50")  # Apply a stipple pattern for transparency
root.mainloop()
```
